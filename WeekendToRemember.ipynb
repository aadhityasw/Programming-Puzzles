This puzzle deals with the problem of inviting friends to dinner over two days
such that no two of your friends who dislike each other are invited on the same
day.<br>This can be done if the graph is a bipartite graph.


```python
def bipartite_graph_color(graph, start, colored_graph, color) :
    if start not in graph :
        return(False, {})
    if start not in colored_graph :
        colored_graph[start] = color
    elif colored_graph[start] != color :
        return(False, {})
    else :
        return(True, colored_graph)
    if color == 'red' :
        color = 'blue'
    elif color == 'blue' :
        color = 'red'
    for i in graph[start] :
        flag, colored_graph = bipartite_graph_color(graph, i, colored_graph, color)
        if flag == False :
            return(False, {})
    return(True, colored_graph)

def color_graph(graph, colored_graph) :
    for i in graph :
        if i not in colored_graph :
            flag, colored_graph = bipartite_graph_color(graph, i, colored_graph, 'red')
            if flag == False :
                return(False, {})
    return(True, colored_graph)

graph = graph7
flag, colored_graph = color_graph(graph, {})
if flag == False :
    print('Such an arrangement is not possible.')
else :
    print('The colored nodes with their respective colors are as follows :- ')
    for i in colored_graph.keys() :
        print(i, '-', colored_graph[i])
```
