# Knight's Tour
The knight is placed on the first block of an empty board and, moving according to the rules of chess, must visit each square exactly once.

```python
def print_Board(board) :
    for i in range(len(board)) :
        for j in range(len(board[i])) :
            print(board[i][j], end = '\t')
        print()
    print()
    return

def make_Move(board, size, pos, index) :
    if index == size*size :
        print_Board(board)
    else :
        moves = [[1, 2], [-1, 2], [2, 1], [-2, 1], [1, -2], [-1, -2], [-2, -1], [2, -1]]
        for move in moves :
            if (pos[0]+move[0] < size) and (pos[0]+move[0] >= 0) and (pos[1]+move[1] < size) and (pos[1]+move[1] >= 0) :
                if board[pos[0]+move[0]][pos[1]+move[1]] == -1 :
                    board[pos[0]+move[0]][pos[1]+move[1]] = index
                    make_Move(board, size, [pos[0]+move[0], pos[1]+move[1]], index+1)
                    board[pos[0]+move[0]][pos[1]+move[1]] = -1
    return

size = int(input('Enter the dimensions of the square shaped board : '))
board = [[-1]*size for i in range(size)]
board[0][0] = 0
print('The various possible combinations are :- ')
make_Move(board, size, [0, 0], 1)
```
