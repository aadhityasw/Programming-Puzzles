# Crossword
A Crossword grid is provided to you, along with a set of words (or names of places) which need to be filled into the grid.  Cells marked with a - are to be filled with the word list.

```python
def check_Validity() :             # To check the validity of the present state of the crossword puzzle.
    global size, board
    for i in range(size) :
        for j in range(size) :
            if board[i][j] == '-' :
                return False
    return True

def make_Move(word, location) :   # To place a word in a particular place in the puzzle.
    global board, size
    x, y, axis = location
    if axis == 0 : # Horizontal
        for i in range(len(word)) :
            board[x][y+i] = word[i]
    elif axis == 1 : #Vertical
        for i in range(len(word)) :
            board[x+i][y] = word[i]
    return

def retrace_Move(word, location) :   # To retract a word that was previously placed in the puzzle.
    global board, size, ch
    x, y, axis = location
    if axis == 0 : # Horizontal
        for i in range(len(word)) :
            if (x==0 or board[x-1][y+i] in [ch, '-']) and (x==(size-1) or board[x+1][y+i] in [ch, '-']) :
                board[x][y+i] = '-'
    elif axis == 1 : #Vertical
        for i in range(len(word)) :
            if (y==0 or board[x+i][y-1] in [ch, '-']) and (y==(size-1) or board[x+i][y+1] in [ch, '-']) :
                board[x+i][y] = '-'
    return

def print_Board() :      # To print the crossword puzzle.
    global board, size
    print('The final solved board is :- ')
    for i in range(size) :
        for j in range(size) :
            print(board[i][j], end='')
        print()
    return

def possible_Locations(word) :   # To find the all possible locations of placing a word in the puzzle.
    global board, size, ch
    possibility = []
    length = len(word)
    for i in range(size) :
        for j in range(size) :
            if board[i][j] != ch :
                poss_Horizontal = True
                poss_Vertical = True
                # Horizontal check
                if board[i][j-1] == ch :
                    k = 0
                    while (k+j) < size and board[i][j+k] != ch :
                        if k >= length :
                            poss_Horizontal = False
                        else :
                            if board[i][j+k] not in ['-',word[k]] :
                                poss_Horizontal = False
                        k += 1
                else :
                    poss_Horizontal = False
                if poss_Horizontal == True and k == length:
                    possibility.append((i, j, 0))
                # Vertical check
                if board[i-1][j] == ch :
                    k = 0
                    while (k+i) < size and board[i+k][j] != ch :
                        if k >= length :
                            poss_Vertical = False
                            break
                        else :
                            if board[i+k][j] not in ['-',word[k]] :
                                poss_Vertical = False
                        k += 1
                else :
                    poss_Vertical = False
                if poss_Vertical == True and k == length:
                    possibility.append((i, j, 1))
    return possibility

def solve(solved, words) :     # The main recursive function that solves the puzzle.
    if check_Validity() == True :
        return True
    if len(words) != 0  :
        global board
        word = words.pop()
        possibility = possible_Locations(word)
        for location in possibility :
            make_Move(word, location)
            ans = solve(solved, words)
            if ans == True :
                solved = True
                return solved
            retrace_Move(word, location)
    return(solved)

def assert_Position(words) :     # To assert the positions of the words if their position in the board is obvious.
    global board
    i = 0
    while i < len(words) :
        word = words[i]
        possibility = possible_Locations(word)
        if len(possibility) == 1 :
            make_Move(word, possibility[0])
            words.remove(word)
            i -= 1
        i += 1
    return(words)

def get_Input() :    # To get the Puzzle from the user.
    global size, board, words
    size = int(input('Enter the size of the crossword grid :- '))
    print('Enter the crossword line by line :- ')
    for i in range(size) :
        s = input()
        board.append([])
        for letter in s :
            board[i].append(letter)
    words = list(input('Enter the words seperated by spaces :- ').strip().split(' '))

board = []
words = []
size = 0
get_Input()
ch = ''
for i in range(size) :
    for j in range(size) :
        if board[i][j] != '-' :
            ch = board[i][j]
            break
    if ch != '' :
        break
words = assert_Position(words)
ans = solve(False, words)
if ans == True :
    print_Board()
```
