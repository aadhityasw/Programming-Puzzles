# The Mating Algorithm
Each Day: <br>
Morning:– <br> Each girl stands on her balcony–Each boy stands under the balcony of his favorite girl whom he has not yet crossed off his list and serenades. If there are no girls left on his list, he stays home and does homework. <br> Afternoon:– <br> Girls who have at least one suitor say to their favorite from among the suitors that day :“Maybe, come back tomorrow.”– To the others, they say “No, I will never marry you!” <br> 
Evening:– <br> Any boy who hears “No” crosses that girl of his list. <br>
Termination Condition: If there is a day when every girl has at most one suitor, we stop and each girl marries her current suitor (if any).


```python
num_female = int(input('Enter the number of female candidates :- '))
female = {}
for i in range(num_female) :
    name = input('Enter the name of the female candidate :- ')
    li = input('Enter all the male candidates that', name, 'likes in their order of preference :- ').strip().split(' ')
    female[name] = li

num_male = int(input('Enter the number of male candidates :- '))
male = {}
for i in range(num_male) :
    name = input('Enter the name of the male candidate :- ')
    li = input('Enter all the female candidates that', name, 'likes in their order of preference :- ').strip().split(' ')
    male[name] = li

crossouts = {}
for i in male.keys() :
    crossouts[i] = []

serenade = {}
for i in female.keys() :
    serenade[i] = []

for i in male.keys() :
    serenade[male[i][0]].append(i)

t = False
while(not t) :
    reject = []
    for i in serenade.keys() : # The Female
        for j in female[i] :
            if j in serenade[i] :
                k = 0
                while(k<len(serenade[i])) :  # The male
                    per = serenade[i][k]
                    if per != j :
                        crossouts[per].append(i)
                        serenade[i].remove(per)
                        reject.append(per)
                    else :
                        k += 1
                break
    
    for i in reject : # The rejected male on that day.
        for j in male[i] : # The list of females that he likes
            if j not in crossouts[i] :
                serenade[j].append(i)
                break
    
    f = False
    for i in serenade.values() :
        if len(i)>1 :
            f = True
            break
    
    if f == False :
        t = True
        break

print('The final list of couples are as follows :- ')
for i in serenade.keys() :
    if len(serenade[i]) > 0 :
        print(i, 'and', serenade[i][0])

# The leftover people who were not paired with anyone.
leftover = []
for i in female.keys() :
    if len(serenade[i]) == 0 :
        leftover.append(i)
for i in male.keys() :
    if [i] not in serenade.values() :
        leftover.append(i)
if len(leftover) != 0 :
    print('The unlucky people in this arrangement are as follows :- ')
    for i in leftover :
        print(i)

# To check for rogue couples.
rogue = []
for i, j in serenade.items() :
    if len(j) > 0 :
        m = j[0]
        if male[m].index(i)>0 :
            for k in range(male[m].index(i)) :
                fe = male[m][k]
                if serenade[fe] is not None and female[fe].index(m) < female[fe].index(serenade[fe][0]) :
                    rogue.append(m, fe)
                    break
if len(rogue) > 0 :
    print('The rogue couples in this setup are :- ')
    for i in rogue :
        print(i[0], 'and', i[1])
else :
    print('There are no rogue couples in this arrangement.')
```
