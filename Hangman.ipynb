```python
import string
import random

#Importing all the valid words from the file.
file_name = open("words.txt",'r')
each_line = file_name.readline()
wordlist=each_line.split()

#To Choose a random word from the list of words.
word=random.choice(wordlist)

#The Game
warn = 3
guess = 6
length=len(word)
wo=[]
correct=False
avalible='abcdefghijklmnopqrstuvwxyz'
entered=''
for i in range(length) :
    wo.append('_')

#Input of a character.
print("Please make yourselves comfortable as the game is about to start.")
print("\t\t Welcome to Hangman")
print('You have',warn,'Warnings left!')
print('You have',guess,'Guesses left!')
print('----------------------------------------------------------------')
print('Your avalible choices are:-',avalible)
ch=input('Enter your guess:')
while (guess!=0) :
    guess=guess-1
    if ch=='*' :
        guess=guess+1
        p=[]
        for i in range(len(wordlist)) :
            e=0
            g=0
            for j in range(len(wo)) :
                if len(wo)==len(wordlist[i]) :
                    if wo[j] != '_' :
                        e=e+1
                        if wo[j] ==wordlist[i][j] :
                            corr=True
                            g=g+1
            if e==g and e!=0:
                p.append(wordlist[i])
        if len(p)==0 :
            print('You cannot ask for hints without making any correct guesses.')
        else :
            print('The hint words are :- ',p)
    elif (not(ch.isalpha())) :
        print('Oops! You have not entered an alphabet.')
        if warn>0 :
            warn=warn-1
            print('You have lost a Warning')
        else :
            guess=guess-1
            print('As all your Warnings have already been exhausted, you have lost a guess.')
            if guess==0 :
                break
    else:
        ch=ch.lower()
        if ch in entered :
            print('You have guessed a letter that you have already entered')
            if warn>0 :
                warn=warn-1
                print('You have lost a Warning')
            else :
                guess=guess-1
                print('As all your Warnings have already been exhausted, you have lost a guess.')
        else :
            a=avalible.index(ch)
            avalible=avalible[:a]+avalible[a+1:]
            if ch not in word :
                print('Oops! Your guess is wrong. The letter',ch,'is not present in the word.')
            else :
                c=False
                for i in range(len(word)) :
                    if word[i]==ch :
                        wo[i]=ch
                        c=True
                if c==True :
                    print('Good Guess! Your guess is correct.')
    print('----------------------------------------------------------------')
    entered=entered+ch
    if str(wo)==word :
        correct=True
        break
    if guess==0 :
        print('You have exhausted all your guesses')
        break
    if correct==False :
        if warn>1 or warn==0:
            print('You have',warn,'Warnings left!')
        elif warn==1 :
            print('You have',warn,'Warning left! Be careful while entering your guess.')
        if guess>1 :
            print('You have',guess,'guesses remaining.')
        elif guess==1:
            print('You have',guess,'guess remaining. Make it count.')
        print('Your Game\'s status is :',str(wo))
        print('Your avalible choices are :-',avalible)
        ch=input('Enter your next guess:');

if warn>1 :
    print('You had',warn,'Warnings left.')
elif warn==1 :
    print('You had',warn,'Warning left.')
if guess>1 :
    print('You had',guess,'guesses remaining.')
elif guess==1:
    print('You had',guess,'guess remaining.')
if correct==True :
    print('Congratulations! You have won the game.')
else :
    print('Sorry! You have lost the game.')
    print('Your status of the game in the end was :-',str(wo))
print('The secret word is :',word)
```
